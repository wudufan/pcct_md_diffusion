```python
import SimpleITK as sitk
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
input_dir = '/fs/ess/PAS3128/pcct/data/omnitom_pcd/recon/img/44'
sitk_img0 = sitk.ReadImage(os.path.join(input_dir, 'img_ch0.nii.gz'))
sitk_img1 = sitk.ReadImage(os.path.join(input_dir, 'img_ch1.nii.gz'))
sitk_img2 = sitk.ReadImage(os.path.join(input_dir, 'img_ch2.nii.gz'))
```

```python
img0 = sitk.GetArrayFromImage(sitk_img0).astype(np.float32)
img0[img0 < -1024] = -1024
img0 = (img0 + 1000) / 1000 * 0.0196

img1 = sitk.GetArrayFromImage(sitk_img1).astype(np.float32)
img1[img1 < -1024] = -1024
img1 = (img1 + 1000) / 1000 * 0.0192

img2 = sitk.GetArrayFromImage(sitk_img2).astype(np.float32)
img2[img2 < -1024] = -1024
img2 = (img2 + 1000) / 1000 * 0.0186
```

```python
diff = img0 - img2
```

```python
plt.imshow(diff[diff.shape[0]//2], cmap='gray', vmin=0.0005, vmax=0.002)
```
